```python
import pandas as pd
import pathlib

way_metadata_dir = pathlib.Path("../data/metadata/idr0080-way-perturbation/screenA/idr0080-way-perturbation_screenA_2701.parquet.gzip")
way_metadata = pd.read_parquet(way_metadata_dir)
print(way_metadata)

num_wells = way_metadata["well_id"].nunique()
num_plates = way_metadata["plate_id"].nunique()
img_methods = way_metadata["imaging_method"].unique()
samples = way_metadata["sample"].unique()
organisms = way_metadata["organism"].unique()
unique_genes = way_metadata["gene_identifier"].unique()
unique_phenotypes = way_metadata["phenotype_identifier"].unique()
stains_targets = way_metadata["stains_targets"].unique()
cell_lines = way_metadata["cell_line"].unique()

print("Num Wells: ", num_wells)
print("Num Plates: ", num_plates)
print("Img Methods: ", img_methods, f"n={len(img_methods)}")
print("Sample Types: ", samples, f"n={len(samples)}")
print("Organisms: ", organisms, f"n={len(organisms)}")
print("Unique Genes: ", unique_genes, f"n={len(unique_genes) - 1}")
print("Unique Phenotypes: ", unique_phenotypes, f"n={len(unique_phenotypes) - 1}")
print("Unique Stains & Targets: ", stains_targets, f"n={len(stains_targets)}")
print("Unique Cell Lines: ", cell_lines, f"n={len(cell_lines)}")
```

```python
from plotnine import *
from sigfig import round

study_imgs_dir = pathlib.Path('/home/parkerhicks/Documents/Prototype_Graphs/Way_Stats')
way_stats_dir = pathlib.Path("/home/parkerhicks/Documents/individual_studies_diversity.parquet.gzip")
way_stats = pd.read_parquet(way_stats_dir)
way_stats = way_stats.loc[way_stats["Study_Name"] == "idr0080-way-perturbation_screenA_2701"]

attributes_to_round = ["H", "J", "NME", "E", "GC"]
for attribute in attributes_to_round:
        way_stats[attribute] = way_stats[attribute].apply(lambda x: round(x, sigfigs=3) if x>0 else x)

dodge_text = position_dodge(width=0.9)

attributes_to_graph = ["H", "J", "NME", "E", "GC", "S"]
for attribute in attributes_to_graph:
        p = (ggplot(data=way_stats, mapping=aes(x="Attribute", y=attribute)) + 
        geom_col(na_rm=True, stat='identity', position='dodge') +
        theme(axis_text_x=element_text(rotation=90)) + 
        geom_text(aes(label=attribute),
                position=dodge_text,
                color='gray', size=8, va='bottom'))

        st_output_file = pathlib.Path(study_imgs_dir, f"{attribute}.png")
        p.save(st_output_file)
```
